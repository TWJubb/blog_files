## Imports

```python
from record_io import *
import pandas as pd
import os
```

```python
#
# SET THE FOLLOWING PATHS CORRECTLY
#

BASE_PATH      = os.getcwd()
RECORD_PATH    = BASE_PATH + os.sep + "records"
PATH_TO_IMAGES = BASE_PATH + os.sep + "images"
PATH_TO_LABELS = BASE_PATH + os.sep + "images\item_map.pbtxt"

test_fraction = 0.1
```

```python
def _class_to_index_mapping():
    category_index = label_map_util.create_category_index_from_labelmap(PATH_TO_LABELS, use_display_name=True)

    class_to_index = {}
    for key, value in category_index.items():
        class_to_index[value["name"]] = key
        
    return class_to_index

def _index_to_class_mapping():
    category_index = label_map_util.create_category_index_from_labelmap(PATH_TO_LABELS, use_display_name=True)

    index_to_class = {}
    for key, value in category_index.items():
        index_to_class[key] = value["name"]
        
    return index_to_class
```

## Example 1 : Create Record File

```python
#---------------------------------------------
# Read in the class labels for the images
#---------------------------------------------

df = pd.read_csv(PATH_TO_IMAGES + os.sep + "labels.txt", 
                 header=None, 
                 names=["File", "xmin", "ymin", "xmax", "ymax", "Label"])
df.head()
```

```python
#---------------------------------------------
# Gather all the image data together
#---------------------------------------------

images_and_bboxes = dict()

for idx, obj in df.iterrows():
    
    box_coords = [obj["xmin"],obj["ymin"],obj["xmax"],obj["ymax"]]
    filename = obj["File"]
    
    if filename not in images_and_bboxes:
        images_and_bboxes[filename] = {"box_coords":[], "class":[]}
        
    images_and_bboxes[filename]["box_coords"] += [box_coords]
    images_and_bboxes[filename]["class"] += [obj["Label"] ]
```

```python
#---------------------------------------------
# Create the mapping from class name -> index
#---------------------------------------------

class_to_index = _class_to_index_mapping()

# check for mistakes
for label in df["Label"].unique():
    if label not in class_to_index:
        print("Class {} not in the labels file \n\t {}".format(label, PATH_TO_LABELS))
```

```python
#---------------------------------------------
# Train test split
#---------------------------------------------

cut = int(np.ceil(len(images_and_bboxes)* test_fraction))
train_examples = dict(list(images_and_bboxes.items())[cut:])
test_examples  = dict(list(images_and_bboxes.items())[:cut])

create_record_file(PATH_TO_IMAGES, RECORD_PATH + os.sep + r'train.record', train_examples, class_to_index)
create_record_file(PATH_TO_IMAGES, RECORD_PATH + os.sep + r'test.record', test_examples, class_to_index)
```
